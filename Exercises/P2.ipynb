# Exercise 2

```python
plt.scatter(df["Sales"], df["TV"])
plt.xlabel("Sales")
plt.ylabel("TV")
plt.show()
```

The scatterplot shows us that there's a trend of "The more you spend, the more you make".

```python
cov = df["Sales"].cov(df["TV"])
corr = df["Sales"].corr(df["TV"])
cov, corr
```

The results:

cov gives us a positive value --> as TV increases, Sales also increase.
corr value --> There is a strong positive linear relationship between TV and Sales.

```python
df.describe()
```

count = number of entries
min = minimum value
max = maximum values
25% = the first percentile of data is lesser or equal to this value
50% = the second percentile of data is lesser or equal to this value and greater than the 25%
75% = the third percentile of data is lesser or equal to this value and greater than the 50%
above the 75% lies the fourth percentile.

```python
mean_sales = df["Sales"].sum() / len(df)
mean_TV = df["TV"].sum() / len(df)
mean_Radio = df["Radio"].sum() / len(df)
mean_Newspaper = df["Newspaper"].sum() / len(df)
mean_sales, mean_TV, mean_Radio, mean_Newspaper
```

```python
std_sales = np.sqrt(((df["Sales"] - mean_sales)**2).sum() / (len(df) - 1))
std_TV = np.sqrt(((df["TV"] - mean_TV)**2).sum() / (len(df) - 1))
std_Radio = np.sqrt(((df["Radio"] - mean_Radio)**2).sum() / (len(df) - 1))
std_Newspaper = np.sqrt(((df["Newspaper"] - mean_Newspaper)**2).sum() / (len(df) - 1))
std_sales, std_TV, std_Radio, std_Newspaper
```

```python
median_sales = df["Sales"].median()
median_TV = df["TV"].median()
median_Radio = df["Radio"].median()
median_Newspaper = df["Newspaper"].median()
median_sales, median_TV, median_Radio, median_Newspaper
```

```python
mode_sales = df["Sales"].mode()[0]
mode_TV = df["TV"].mode()[0]
mode_Radio = df["Radio"].mode()[0]
mode_Newspaper = df["Newspaper"].mode()[0]
mode_sales, mode_TV, mode_Radio, mode_Newspaper
```

Median and Mode compared to Mean:
Median picks out the number in the middle of the dataseries. Mode picks out the mose frequent numeber. Mean adds all the numbers together and devides it by the amount of numbers, resulting in an average number.

```python
s2_sales = ((df["Sales"] - mean_sales)**2).sum() / (len(df) - 1)
s2_tv = ((df["TV"] - mean_TV)**2).sum() / (len(df) - 1)
s2_radio = ((df["Radio"] - mean_Radio)**2).sum() / (len(df) - 1)
s2_newspaper = ((df["Newspaper"] - mean_Newspaper)**2).sum() / (len(df) - 1)
s2_sales, s2_tv, s2_radio, s2_newspaper
```

```python
var_sales = df["Sales"].var()
var_TV = df["TV"].var()
var_Radio = df["Radio"].var()
var_Newspaper = df["Newspaper"].var()
var_sales, var_TV, var_Radio, var_Newspaper
```

Computing sample variance on my own and using pandas and comparing the results shows that pandas indeed uses Bessel's correction.
Bessel's correction is used because we do not know the population. this way we make an unbiased calculation
