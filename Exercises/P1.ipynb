# Random sampling from statistical distributions

```python
rng = np.random.default_rng()

X_int = rng.integers(100, size=100)
X_int
```

```python
X_uniform = rng.uniform(size=100)
X_uniform
```

```python
X_normal = rng.normal(size=100)
X_normal
```

```python
X_binomial = rng.binomial(n=10, p=.3, size=100)
X_binomial
```

```python
X_neg_binomial = rng.negative_binomial(n=10, p=.3, size=100)
X_neg_binomial
```

```python
X_gamma = rng.gamma(shape=1.0, scale=0.9, size=100)
X_gamma
```

```python
X_geom = rng.geometric(p=0.31, size=100)
X_geom
```

# Exercise 1

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(3, 3, figsize = (12, 12))
arrays = [X_int, X_uniform, X_normal, X_binomial, X_neg_binomial, X_gamma, X_geom]
axes = axes.flatten()
titles = ["Uniform Integer", "Uniform", "Normal", "Binomial", "Negative Binomial", "Gamma", "Geometric"]

for ax, arr, title in zip(axes, arrays, titles):
    counts, bin_edges = np.histogram(arr, bins=20)
    sns.histplot(arr, bins=20, ax=ax, alpha=0.6)
    ax.set_title(title)

    cumulative = np.cumsum(counts)
    bin_centers = (bin_edges[:-1] + bin_edges[1:])/2
    ax2 = ax.twinx()
    ax2.plot(bin_centers, cumulative, marker='o', color = 'red')
    ax.set_ylabel("Frequency")
    ax2.set_ylabel("Cumulative Frequency")

for ax in axes[len(arrays):]:
    ax.remove()

plt.tight_layout()
plt.show()
```
